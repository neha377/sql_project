## Information about all the columns in the dataset
1. **Item_Identifier:** This column contains unique identifiers for items, such as product codes or IDs.

2. **Item_Weight:** The weight of products.

3. **Item_Fat_Content:** The fat content of the items.

4. **Item_Visibility:** The items on display in a store, in numerical value.

5. **Item_Type:** The contains the type or category of each item.

6. **Item_MRP:** The Maximum Retail Price (MRP) of the items.

7. **Outlet_Identifier:** The unique identifiers for different outlets.

8. **Outlet_Establishment_Year:** The year in which each outlet was established or opened.

9. **Outlet_Size:** This column contains information about the size of the outlets.

10. **Outlet_Location_Type:** It contains the location type of each outlet, as Tier 1, Tier 2, Tier 3.

11. **Outlet_Type:** This column describes the type of outlet or store.

12. **Item_Outlet_Sales:** This column contains the sales figures for each item.
